```python
from src.text_clustering import ClusterClassifier
from datasets import load_dataset
import os
os.environ['CURL_CA_BUNDLE'] = ''

texts = load_dataset('openpecha/cleaned_MT_v1.0.2', split="train")["Target"]

def batch_and_join(strings, n):
    # Create batches of 'n' strings joined with new lines
    batched_strings = ["\n".join(strings[i:i + n]) for i in range(0, len(strings), n)]
    return batched_strings

batched_texts = batch_and_join(texts, 20)

cc = ClusterClassifier()
```

```python
cc.fit(batched_texts, clustering_algorithm='kmeans')
```

```python
cc.show()
```
